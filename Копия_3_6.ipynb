```python
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def f(x):
    return (np.sin(x + 1))

x0 =  fsolve(f, 1)
print (x0)

x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, f(x))
plt.plot([-10, 10],[0,0], color ='red', linewidth=0.75, linestyle ="--") 
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
